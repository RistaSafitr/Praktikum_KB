```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

Melihat data, tipe data, dan memeriksa data null dari dataset yg digunakan

```python
df_ranking = pd.read_csv("2023 QS World University Rankings.csv")
df_ranking.head(50)
```

```python
df_ranking.info()
```

```python
df_ranking.isnull().sum()
```

Grafik batang - Melihat lokasi 20 universitas terbaik

```python
x = df_ranking['location code'].head(20)
y = df_ranking['score scaled'].head(20)
```

```python
z = '#5f857e'
plt.bar(x,y, align = 'center', width = 0.7, color = z)
plt.title('World University Rankings 2023')
plt.xlabel('Lokasi')
plt.ylabel('Score Scaled')
plt.show()
```

Boxplot - terdapat data outlier

```python
sns.boxplot(df_ranking['ar score'])
plt.title('Skor AR')
plt.show()
```

scatter plot - persebaran data 100 peringkat universitas terbaik beerdasarkan lokasinya

```python
sns.scatterplot(data=df_ranking, x = df_ranking['Rank'].head(100), y = df_ranking['location code'])
plt.title('Persebaran 100 Peringkat Teratas Universitas di Dunia ')
plt.xlabel('Peringkat')
plt.ylabel('Lokasi')
plt.show()
```

Grafik Heatmap

```python
warna = sns.light_palette("#1d544a")
sns.heatmap(df_ranking.corr(), annot=True, cmap = warna)
```

pie chart - semua perbandingan data

```python
df_pie = df_ranking['location code'].value_counts().rename_axis('location_code').reset_index(name='ger_rank')
```

```python
plt.figure(figsize=(10,10))
plt.pie(df_pie.ger_rank, labels=df_pie.location_code, startangle=90,
autopct='%.1f%%')
plt.title('Presentase Universitas Terbaik di Dunia')
plt.show()
```

pie chart - perbandingan letak lokasi 50 universitas peringkat teratas 

```python
y = df_ranking['location code']
x = df_ranking['Rank'].head(50)
```

```python
df_pie = x.value_counts().rename_axis('location_code').reset_index(name='Rank')
```

```python
plt.figure(figsize=(10,10))
plt.pie(df_pie.Rank, labels=df_pie.location_code, startangle=90,
autopct='%.1f%%')
plt.title('Presentase Lokasi 50 Universitas Terbaik di Dunia')
plt.show()
```
