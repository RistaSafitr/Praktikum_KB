K-Means Clustering Algorithm

Import library

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

Membaca dataset

```python
df = pd.read_csv("Nasa_neo_2022.csv")
```

```python
df.head(10)
```

Membuat objek dari k means dengan jumlah kluster 5

```python
neo_model = KMeans(n_clusters = 5)
```

Memilih atribut yang akan dilatih

```python
points = df[["relative_velocity", "miss_distance"]]
```

Melakukan fit terhadap neo_model

```python
neo_model.fit(points)
```

Menentukan label untuk cluster

```python
labels = neo_model.labels_
```

Mencetak label

```python
print(labels)
```

Membuat dataframe dengan label dan nama sebagai kolom

```python
neo_KMeans = pd.DataFrame({'labels': labels, 'name': df.name})
```

Membuat crosstab

```python
crosstab = pd.crosstab(neo_KMeans["labels"], neo_KMeans["name"])
```

Mencetak crosstab

```python
print(crosstab)
```

Mengambil semua baris pada kolom yang telah dipilih sebelumnya

```python
neo_xs = points.iloc[:,0]
neo_ys = points.iloc[:,1]
```

Menampung koordinat dari tiap centorid

```python
centroids = neo_model.cluster_centers_
```

```python
centroids_x = centroids[:,0]
centroids_y = centroids[:,1]
```

Menampilkan visualisasi cluster

```python
plt.figure(figsize=(20, 20))
plt.scatter(neo_xs, neo_ys, alpha = 0.5, c = labels)
plt.scatter(centroids_x, centroids_y, marker = "P", s = 50, label = "Centroid")
plt.xlabel("relative_velocity")
plt.ylabel("miss_distance")
plt.legend()
plt.show()
```

Menentukan jumlah cluster (Elbow Method)

```python
kluster_range = range(1,9)
nasa_found = []

for k in kluster_range:
    neo_model = KMeans(n_clusters = k)
    neo_model.fit(points)
    nasa_found.append(neo_model.inertia_)
```

Menampilkan visualisasi plot kluster_range dengan nasa_found

```python
plt.figure(figsize=(20,20))
plt.plot(kluster_range, nasa_found, '-o')
plt.xlabel('Jumlah Cluster, k')
plt.ylabel('inertia')
plt.xticks(kluster_range)
plt.show()
```
