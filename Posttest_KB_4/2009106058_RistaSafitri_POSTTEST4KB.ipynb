Melakukan Data cleaning pada data dengan nilai null dengan mengganti nilai null (bilangan bulat : median)

```python
df_rusak = pd.read_csv("mushroom_rusak.csv")
```

```python
df_rusak.isnull().sum()
```

```python
df_rusak.shape
```

```python
imputer_median = SimpleImputer(strategy='median')
```

```python
df_rusak_cleaned = df_rusak.copy()
```

```python
df_rusak_cleaned.columns
```

```python
df_rusak_cleaned['RingNumber'] = imputer_median.fit_transform(df_rusak[['RingNumber']])
```

```python
data_pembanding = pd.DataFrame(
    {
        "RingNumber_original":df_rusak['RingNumber'],
        "RingNumber_cleaned":df_rusak_cleaned['RingNumber'],
    }
)
```

```python
data_pembanding.head(30)
```
